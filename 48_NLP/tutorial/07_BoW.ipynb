# Bag of Words Practicals

Bag of Words (BoW) is a simple and widely used technique in Natural Language Processing (NLP) to represent text data as numerical features. This method transforms text into a fixed-length vector of word occurrences.

```python
# import necessary modules
import pandas as pd
```

```python
data = pd.read_csv('spam.csv', encoding='latin-1')
data.head()
```

```python
# Preparing dataset for Data Cleaning
data.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis=1, inplace=True)
data.rename(columns={'v1': 'label', 'v2': 'messages'}, inplace=True)
data.head()
```

```python
# Data Cleaning and Preprocessing
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
nltk.download('stopwords')
nltk.download('wordnet')

Lemmatizer = WordNetLemmatizer()
```

```python
corpus = []

for i in range (0, len(data)):
    review = re.sub('[^a-zA-Z]', ' ', data['messages'][i])
    review = review.lower()
    review = review.split()
    review = [Lemmatizer.lemmatize(word, pos='n') for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)

corpus[0:10]
```

```python
## Create the bag of words
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=2500, binary=False)
X = cv.fit_transform(corpus).toarray()
print(X.shape)
print(X[0])
```
